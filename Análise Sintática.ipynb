# Análise Sintática utilizando SpaCy, junto à biblioteca Displacy.



Análise Sintática: Frase destinada à redução do volume de uma televisão.

```python
import spacy
from spacy import displacy

nlp = spacy.load("pt_core_news_sm")

television_1 = ("Não consigo te ouvir, a TV tá muito alta.")

doc = nlp(television_1)
displacy.render(doc, style = "dep", jupyter = True)
```

Análise Sintática: Frase destinada a desligar o ventilador ou o dispositivo de refrigeração ambiente.

```python
import spacy
from spacy import displacy

nlp = spacy.load("pt_core_news_sm")

fan_2 = ("Tá ficando frio demais.")

doc = nlp(fan_2)
displacy.render(doc, style = "dep", jupyter = True)
```

Pode-se observar que a maioria das setas da árvore partem a partir dos verbos. Portanto, pode ser uma boa ideia começarmos a observar as relações na árvore a partir deles. Verbos se ligam a advérbios, pronomes, adjetivos e substantivos.

Avaliando primariamente os substantivos, pode-se ver, diretamente, que nesse caso, eles são o objeto central da frase, ou seja, o objeto a ser manipulado pelo interpretador de textos. Os verbos devem ser avaliados, posteriormente, em vias de se verificar a ação a ser tomada.

Em alguns casos, o que será  indicado não será direto como o comando 'Acender Luz', por exemplo. Como no 1º exemplo que se refere ao ventilador, o usuário somente reclamaria do calor da sala, em vias de que se pudesse ligar o ventilador automaticamente, por exemplo.
